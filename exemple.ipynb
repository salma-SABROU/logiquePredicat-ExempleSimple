```python
import pandas as pd
data = pd.read_csv("zoo.csv")
data
```

```python
data2 = pd.read_csv("class.csv")
data2
```

```python
#En teste par l'entite animaux
animaux=data["animal_name"].values

animaux
```

```python
# predicat : est_aminal
def est_animal(animal):
    return animal in animaux

#predicat : est_mammifere
def est_mammifere(animal):
    mammiferes = data2.loc[data2['Class_Type'] == 'Mammal', 'Animal_Names'].tolist()
    return animal in mammiferes

#predicat : est_carnivore
def est_carnivore(animal):
    carnivore_classes = [1, 2, 3, 4, 5, 7]
    animal_class = data.loc[data["animal_name"] == animal, "class_type"].values[0]
    return animal_class in carnivore_classes
    
```

```python
# Règle : regle_est_carnivore_mammifere(X) : - est_animal(X),est_mammifere(X),est_carnivore(X). 
#Cette règle signifie que si X un animal est à la fois un mammifère et carnivore, alors il est un mammifère carnivore.
def regle_est_carnivore_mammifere(animal):
    if est_animal(animal):
        return est_mammifere(animal) and est_carnivore(animal)
    else:
        return False
    
```

```python
#le fait :nous pouvons définir le fait "le lion est carnivore" en utilisant la colonne "predator" qui indique 
#sil'animal est un prédateur ou non. Si la valeur de la colonne pour le lion est 1, cela signifie que le lion est un prédateur, donc le fait est vrai. Si la valeur est 0, le lion n'est pas un prédateur, donc le fait est faux.
est_predator = lambda animal: data.loc[data["animal_name"] == animal, "predator"].values[0] == 1
print("antelope",est_predator("antelope"))
print("lion : ",est_predator("lion"))
```
